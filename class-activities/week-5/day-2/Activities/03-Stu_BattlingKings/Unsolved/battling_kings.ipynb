```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
df = pd.read_csv("Resources/got.csv")
df.head()
```

```python
# Get attacker and defender data
df_atk_def = df[["attacker_king", "defender_king"]]
df_atk_def
```

```python
# Get total battle data
df_battle = pd.DataFrame({"attack": df_atk_def["attacker_king"].value_counts(), "defend" : df_atk_def["defender_king"].value_counts()}).fillna(0)
df_battle["Total"] = df_battle["attack"] + df_battle["defend"]
df_battle
```

```python
# Configure plot and ticks
df_battle[["Total"]].plot(kind = "bar", figsize = (12,10), rot = 45, color = "r")
```
